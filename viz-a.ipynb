# Data visulization-a

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
 car_crashes = sns.load_dataset('car_crashes').head(5)
```

```python
car_crashes
```

```python
sns.relplot(x='alcohol', y=
'total', data=car_crashes)
```

```python
sns.lmplot(x='alcohol', y='total', data= car_crashes )
```

```python
sns.barplot(x='alcohol', y='total', data= car_crashes )
```

#### conclusion
 There is a posivtive relationship between alcohol and the total car crashes
